```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the parquet file
df_phase = pd.read_parquet('umich_2021_phase.parquet')
df_time = pd.read_parquet('umich_2021_time.parquet')
df_phase.columns
```

```python

# Z is actually -Y
# X is actually -X
# Y is actually Z

start = 1000
end = start + 500


# Plot the data for phase and time side to sid
fig, ax = plt.subplots(1, 2, figsize=(10, 10))
for axis in ['x', 'y', 'z']:
    ax[0].plot(df_phase.loc[start:end,f'grf_{axis}_l'], label=f'GRF {axis.upper()}')
    ax[1].plot(df_time.loc[start:end,f'grf_{axis}_l'], label=f'GRF {axis.upper()} time')


plt.xlabel('Time')
plt.ylabel('GRF')
ax[0].legend()
ax[1].legend()
```

```python
# Plot CoP
fig, ax = plt.subplots(1, 2, figsize=(10, 10))
for axis in ['x', 'y', 'z']:
    ax[0].plot(df_phase.loc[start:end,f'cop_{axis}_l'], label=f'CoP {axis.upper()}')
    ax[1].plot(df_time.loc[start:end,f'cop_{axis}_l'], label=f'CoP {axis.upper()} time')

plt.xlabel('Time')
plt.ylabel('COP')
ax[0].legend()
ax[1].legend()
```
